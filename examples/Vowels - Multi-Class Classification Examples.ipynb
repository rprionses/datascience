```python
import pandas as pd

vowel_train = pd.read_csv('vowel.train.csv', sep=',',header=0)
vowel_test = pd.read_csv('vowel.test.csv', sep=',',header=0)

print("vowel_test:")
print(vowel_test.head())
print("vowel_train:")
print(vowel_train.head())

y_train = vowel_train.iloc[:,1]
X_train = vowel_train.iloc[:,2:]

y_test = vowel_test.iloc[:,1]
X_test = vowel_test.iloc[:,2:]

print("y_train:")
print(y_train)

print("X_train:")
print(X_train)

print("y_test:")
print(y_test)

print("X_test:")
print(X_test)
```

```python
import sklearn as sk
from sklearn.linear_model import LogisticRegression
from sklearn import svm
from sklearn.ensemble import RandomForestClassifier
from sklearn.neural_network import MLPClassifier

LR = LogisticRegression(random_state=0, solver='lbfgs', multi_class='multinomial', max_iter=1000).fit(X_train, y_train)
LR.predict(X_test)
print(f"LR score: {round(LR.score(X_test, y_test), 4)}")

SVM = svm.SVC(decision_function_shape="ovo").fit(X_train, y_train)
SVM.predict(X_test)
print(f"SVM score: {round(SVM.score(X_test, y_test), 4)}")

RF = RandomForestClassifier(n_estimators=1000, max_depth=10, random_state=0).fit(X_train, y_train)
RF.predict(X_test)
print(f"RF score: {round(RF.score(X_test, y_test), 4)}")

NN = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(150, 10), random_state=1, max_iter=10000).fit(X_train, y_train)
NN.predict(X_test)
print(f"NN score: {round(NN.score(X_test, y_test), 4)}")
```
